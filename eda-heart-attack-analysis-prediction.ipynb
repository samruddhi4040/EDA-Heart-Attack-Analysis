>**Cleaning the data.**

```python
data.isnull().sum()
```

>**Dropping the redundant (unecessary) data.**

```python
report=data.drop(['cp','exng'],axis=1)
```

>**Reprinting the dataset after removing the unecessary data.**

```python
report.head()
```

>**Relationship analysis.**

```python
numeric_columns = report.select_dtypes(include=['number']).columns
corelation=report[numeric_columns].corr()
print(corelation)
```

>**Generating a heatmap on the present data.**

```python
sns.heatmap(corelation,xticklabels=corelation.columns,yticklabels=corelation.columns,annot=True)
```

>**Pairplot.**

```python
sns.pairplot(report)
```

>**Relational plot.**

```python
sns.relplot(x='age',y='chol',hue='sex',data=report)
```

>**Histogram.**

```python
sns.displot(report['thalachh'],bins=5)
```

>**Box plot.**

```python
sns.catplot(x='trtbps',kind='box',data=report)
```
